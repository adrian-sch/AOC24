```python
lines = []

with open("input.txt") as file:
    for line in file:
        lines.append(line.strip())

lines
```

```python
lines_down = []

for i in range(len(lines[0])):
    down = ''

    for j in range(len(lines)):
        down += lines[j][i]
    
    lines_down.append(down)

lines_down
```

```python
lines_diagonal_1 = []
lines_diagonal_2 = []

# Get diagonals
n = len(lines)


for i in range(2*n):
    diag = ''
    for j in range(0,max(i,n)):
        if i-j >= 0 and i-j < n and j >= 0 and j < n:
            diag+=lines[i-j][j]
    if len(diag) > 0:
        lines_diagonal_1.append(diag)



for i in range(-n, n):
    diag = ''
    for j in range(0,max(i,n)):
        if i+j >= 0 and i+j < n and j >= 0 and j < n:
            diag+=lines[i+j][j]
    if len(diag) > 0:
        lines_diagonal_2.append(diag)

lines_diagonal_1
```

```python
import re


def count_all_overlapping(pattern, string):
    matches = []
    pos = 0
    while True:
        match = re.search(pattern, string[pos:])
        if not match:
            break
        matches.append(match.group())
        pos += match.start() + 1  # Move one position ahead to find overlapping matches
    return len(matches)

def count_in_lines(pattern, lines):
    return sum(count_all_overlapping(pattern, line) for line in lines)


pattern = r'XMAS|SAMX'

ans = 0

ans += count_in_lines(pattern, lines)
ans += count_in_lines(pattern, lines_down)
ans += count_in_lines(pattern, lines_diagonal_1)
ans += count_in_lines(pattern, lines_diagonal_2)

ans
```
