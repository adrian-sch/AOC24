```python
rules = []
pages = []

read_rules = True
with open("input.txt") as file:
    for line in file:
        if line == "\n":
            read_rules = False
            continue
        if read_rules:
            rules.append(list(map(int, line.strip().split("|"))))
        else:
            pages.append(list(map(int, line.strip().split(','))))

rules
```

```python
pages
```

```python
ans = 0
for line in pages:
    valid = True
    for rule in rules:
        if rule[0] in line and rule[1] in line:
            if line.index(rule[0]) > line.index(rule[1]):
                valid = False
                break
    if valid:
        ans += line[int(len(line)/2)]

ans
    
```

```python
def is_valid(line, rules):
    valid = True
    for rule in rules:
        if rule[0] in line and rule[1] in line:
            if line.index(rule[0]) > line.index(rule[1]):
                valid = False
                break
    return valid

ans = 0
for line in pages:
    sorted = is_valid(line, rules)

    if sorted:
        continue

    # get relevant rules
    rel_rules = [rule for rule in rules if rule[0] in line and rule[1] in line]

    while not sorted:
        for rule in rel_rules:
            if line.index(rule[0]) > line.index(rule[1]):
                # swap
                line[line.index(rule[0])], line[line.index(rule[1])] = line[line.index(rule[1])], line[line.index(rule[0])]
                break
        sorted = is_valid(line, rules)
    
    ans += line[int(len(line)/2)]

ans
    
```
