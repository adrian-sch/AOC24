```python
grid = []

with open("input.txt") as file:
    for line in file:
        grid.append(line.strip())

grid
```

```python
antennas = {}

for i in range(len(grid)):
    for j in range(len(grid[i])):
        if grid[i][j] != ".":
            if not grid[i][j] in antennas:
                antennas[grid[i][j]] = []
            antennas[grid[i][j]].append((i, j))

antennas
```

```python
def in_grid(pos, grid):
    return 0 <= pos[0] < len(grid) and 0 <= pos[1] < len(grid[0])

def add_pos(pos1, pos2):
    return (pos1[0] + pos2[0], pos1[1] + pos2[1])

def subtract_pos(pos1, pos2):
    return (pos1[0] - pos2[0], pos1[1] - pos2[1])

def get_antinodes_part1(pos1, pos2, grid):
    antinode = []

    antinode1 = add_pos(pos1, subtract_pos(pos1, pos2))
    if in_grid(antinode1, grid):
        antinode.append(antinode1)
    antinode2 = add_pos(pos2, subtract_pos(pos2, pos1))
    if in_grid(antinode2, grid):
        antinode.append(antinode2)
    
    return antinode

def get_antonodes_pos_delta(pos, delta, grid):
    n = 0
    antinode = []
    while True:
        antinode_pos = add_pos(pos, (n * delta[0], n * delta[1]))
        if not in_grid(antinode_pos, grid):
            break
        antinode.append(antinode_pos)
        n += 1
    return antinode

def get_antinodes_part2(pos1, pos2, grid):
    antinodes = []
    delta = subtract_pos(pos1, pos2)
    n = 0
    # from pos1
    antinodes.extend(get_antonodes_pos_delta(pos1, delta, grid))
    # from pos2
    antinodes.extend(get_antonodes_pos_delta(pos2, delta, grid))
    
    return antinodes

def print_antinodes(antinodes, grid):
    for i in range(len(grid)):
        for j in range(len(grid[i])):
            if (i, j) in antinodes and grid[i][j] == ".":
                print("#", end="")
            else:
                print(grid[i][j], end="")
        print()
```

```python
ans = 0

antinodes = []

for antenna in antennas:
    for location in antennas[antenna]:
        for other_location in antennas[antenna]:
            if location != other_location:
                delta = (other_location[0] - location[0], other_location[1] - location[1])
                
                new_antonodes = get_antinodes_part1(location, other_location, grid)

                for new_antinode in new_antonodes:
                    if new_antinode not in antinodes:
                        ans +=1
                        antinodes.append(new_antinode)


ans
```

```python
print_antinodes(antinodes, grid)
```
