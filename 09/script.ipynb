```python
input = open("input.txt").read().strip()

len(input)
```

```python

disc = []

files = []
free_space = []

for idx, number in enumerate(input):
    if idx % 2 == 0:
        file = {
            "id" : len(files),
            "size" : int(number),
        }
        files.append(file)
    else:
        free_space.append(int(number))

files, free_space
```

```python
disc = []

for idx, number in enumerate(input):
    if idx % 2 == 0:
        disc.extend( [files[int(idx/2)]["id"]] * files[int(idx/2)]['size'])
    else:
        disc += '.' * int(number)
        
disc
```

```python

import copy

disc = []
files_c = copy.deepcopy(files)

for idx, number in enumerate(input):
    if idx % 2 == 0:
        disc.extend( [files_c[0]["id"]] * files_c[0]['size'])
        files_c.pop(0)
    else:
        
        while True:
            if len(files_c) == 0:
                break
            elif int(number) <= files_c[-1]['size']:
                disc.extend([files_c[-1]["id"]] * int(number))
                files_c[-1]['size'] -= int(number)
                break
            else:
                disc.extend([files_c[-1]["id"]] * files_c[-1]['size'])
                number = int(number) - files_c[-1]['size']
                files_c.pop()
    if len(files_c) == 0:
        break
        
disc
```

```python
ans = 0

for idx, num in enumerate(disc):
    if num == '.':
        continue
    ans += idx * int(num)
    
ans
```

```python
disc = []
files_c = copy.deepcopy(files)
free_space_c = copy.deepcopy(free_space)
moved_files = copy.deepcopy(files)
    
for i in range(len(files_c) - 1, 0, -1):
    moved_id = moved_files.index(files_c[i])
    for j in range(moved_id):    
        if files_c[i]['size'] <= free_space_c[j]:
            
            free_space_c[j] -= files_c[i]['size']
            if moved_id < len(free_space_c):
                free_space_c[moved_id-1] = free_space_c[moved_id-1] + files_c[i]['size'] + free_space_c[moved_id]
                free_space_c.pop(moved_id)
            else:
                free_space_c.pop(moved_id-1)
            # move file to j
            free_space_c.insert(j, 0)
            
            moved_files.remove(files_c[i])
            moved_files.insert(j+1, files_c[i])
            
            break

    print(i, files_c[i])
    
moved_files
```

```python
disc = []

for i in range(len(moved_files)):
    disc.extend( [moved_files[i]["id"]] * moved_files[i]['size'])
    if i < len(free_space_c):
        disc.extend('.' * free_space_c[i])
        
ans = 0

for idx, num in enumerate(disc):
    if num == '.':
        continue
    ans += idx * int(num)
        
ans, disc
```
