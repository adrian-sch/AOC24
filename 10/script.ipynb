```python
grid = []

with open("input.txt", "r") as f:
    for line in f:
        grid.append([int(char) for char in line.strip()])

grid
```

```python
rows = len(grid)
collumns = len(grid[0])

trailheads = []
for i in range(rows):
    for j in range(collumns):
        if grid[i][j] == 0:
            trailheads.append((i,j))

trailheads
```

```python

dirs = [(0,1),(1,0),(0,-1),(-1,0)]

def walk(i,j,grid):

    ends = []

    if grid[i][j] == 9:
        ends.append((i,j))
        return ends

    for dir in dirs:
        if i+dir[0] < 0 or i+dir[0] >= rows or j+dir[1] < 0 or j+dir[1] >= collumns:
            # out of map
            continue 
        if grid[i+dir[0]][j+dir[1]] == grid[i][j] + 1:
             new_ends = walk(i+dir[0],j+dir[1],grid)
             for new_end in new_ends:
                 if new_end not in ends:
                    ends.append(new_end)

    return ends
        


ans = 0
for trailhead in trailheads:
    i,j = trailhead
    ends = walk(i,j,grid)
    print(ends)
    ans += len(ends)

ans   
```
