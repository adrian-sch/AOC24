```python
stones = []

with open("input.txt", "r") as f:
    for line in f:
        stones = [int(char) for char in line.strip().split(" ")]

stones

for blink in range(25):
    new_stones = stones[:]
    stones_added = 0
    for i in range(len(stones)):
        stone = stones[i]
        if stone == 0:
            new_stones[i + stones_added] = 1
        elif len(str(stone)) % 2 == 0:
            length = len(str(stone))
            new_stones[i + stones_added] = int(str(stone)[:length//2])
            new_stones.insert(i+1 + stones_added, int(str(stone)[length//2:]))
            stones_added += 1
        else:
            new_stones[i + stones_added] = stone * 2024
    stones = new_stones
    # print("stones after blink", blink+1, ":", stones)
    print("number of stones after blink", blink+1, ":", len(stones))
```

```python
from collections import Counter

stones = []

with open("input.txt", "r") as f:
    for line in f:
        stones = [int(char) for char in line.strip().split(" ")]

stones

counter = Counter()
for stone in stones:
    counter[stone] += 1
counter

for blink in range(100000):
    new_counter = Counter()
    for stone, count in counter.items():
        if stone == 0:
            new_counter[1] += count
        elif len(str(stone)) % 2 == 0:
            new_counter[int(str(stone)[:len(str(stone))//2])] += count
            new_counter[int(str(stone)[len(str(stone))//2:])] += count
        else:
            new_counter[stone * 2024] += count
    counter = new_counter
    # print("stones after blink", blink+1, ":", stones)
    print("number of stones after blink", blink+1, ":", sum(counter.values()))
```
