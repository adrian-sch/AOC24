```python
grid = []

grid = open("input.txt", "r").read().splitlines()

rows = len(grid)
cols = len(grid[0])

rows, cols, grid
```

```python
from collections import deque

def is_edge(pos, dir, grid):
    return not(0 <= pos[0] + dir[0] < len(grid) and # not in bounds
            0 <= pos[1] + dir[1] < len(grid[0]) and
            grid[pos[0]+dir[0]][pos[1]+dir[1]] == grid[pos[0]][pos[1]]) # not same field

visited = set()
fields = []

dirs = [(0,1), (-1,0), (0,-1), (1,0)]
diags = [(1,1), (1,-1), (-1,1), (-1,-1)]

for i in range(rows):
    for j in range(cols):
        if (i,j) not in visited:

            queue = deque()
            queue.append((i,j))
            field = {
                "id" : grid[i][j],
                "area": 0,
                "perimeter": 0,
                "sides": 0
            }
            while True:

                if len(queue) == 0:
                    break

                ii, jj = queue.popleft()
                visited.add((ii,jj))
                field["area"] += 1
                for dir in dirs:
                    if is_edge((ii,jj), dir, grid):
                        field["perimeter"] += 1
                    else:
                        if (ii+dir[0], jj+dir[1]) not in visited and (ii+dir[0], jj+dir[1]) not in queue:
                            queue.append((ii+dir[0], jj+dir[1]))

                # check outer corners
                for k in range(4):
                    if is_edge((ii,jj), dirs[k], grid) and is_edge((ii,jj), dirs[(k+1)%4], grid):
                        field["sides"] += 1
                # check inner corners
                for k in range(4):
                    diag = diags[k]
                    edge1 = (diag[0], 0)
                    edge2 = (0, diag[1])

                    if (not(is_edge((ii,jj), edge1, grid) or is_edge((ii,jj), edge2, grid))
                        and is_edge((ii,jj), diag, grid)):
                            field["sides"] += 1

            fields.append(field)

ans_1 = 0
ans_2 = 0

for field in fields:
    print(field)
    ans_1 += field["area"] * field["perimeter"]
    ans_2 += field["area"] * field["sides"]

ans_1, ans_2
```
