```python
def parse_button(str) -> dict():
    input_str = str.strip().split(" ")
    input = {}
    button = input_str[1][:-1]
    input[button] = {}
    for i in input_str:
        if "X" in i:
            input[button]["X"] = int(i.split("+")[-1][:-1])
        if "Y" in i:
            input[button]["Y"] = int(i.split("+")[-1])
    return input
            
def parse_prize(str) -> dict():
    input_str = str.strip().split(" ")
    input["P"] = {}
    for i in input_str:
        if "X" in i:
            input["P"]["X"] = int(i.split("=")[-1][:-1]) + 10000000000000
        if "Y" in i:
            input["P"]["Y"] = int(i.split("=")[-1]) + 10000000000000
    return input

inputs = []

with open("input.txt", "r") as f:
    input = {}
    for line in f:
        if line == "\n":
            inputs.append(input)
            input = {}
        elif "A" in line or "B" in line:
            input.update(parse_button(line))
        elif "Prize" in line:
            input.update(parse_prize(line))
    inputs.append(input)

inputs
```

```python

def check_integer(x):
    tolerance = 1e-2
    return abs(x - round(x)) < tolerance

ans = 0

for input in inputs:
    
    button_a = (input["P"]["Y"] - input["B"]["Y"] * (input["P"]["X"]/input["B"]["X"]))/(input["A"]["Y"] - input["B"]["Y"] * (input["A"]["X"]/input["B"]["X"]))
    button_b = (input["P"]["X"] - input["A"]["X"] * button_a)/input["B"]["X"]
    
    print(button_a, button_b)
    if check_integer(button_a) and check_integer(button_b) and button_a >= 0 and button_b >= 0:
        ans += button_a * 3 + button_b
    
ans
```
