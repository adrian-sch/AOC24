```python
grid = open('input.txt').read().splitlines()

rows = len(grid)
cols = len(grid[0])

# start and end are always in the corners            
start = (rows - 2, 1)
end = (1, cols - 2)

rows, cols, start, end
```

```python
from collections import deque

def in_grid(pos, grid):
    return 0 <= pos[0] < len(grid) and 0 <= pos[1] < len(grid[0])

def is_wall(pos, grid):
    return grid[pos[0]][pos[1]] == "#"

def is_end(pos):
    return pos == end

def is_empty(pos, grid):
    return grid[pos[0]][pos[1]] == "."

def walk(pos, dir, path, score, grid, visited):
    ret = []
    new_pos = (pos[0] + dir[0], pos[1] + dir[1])
    if (is_empty(new_pos, grid) and
        ((new_pos, dir) not in visited or visited[(new_pos, dir)] >= score + 1) 
         ) or is_end(new_pos):
        new_path = path.copy()
        new_path.append(new_pos)
        ret.append((new_pos, dir, new_path, score + 1))
    
    turn_cw = (dir[1], -dir[0])
    new_pos = (pos[0] + turn_cw[0], pos[1] + turn_cw[1])
    if (is_empty(new_pos, grid) and
        ((new_pos, turn_cw) not in visited or visited[(new_pos, turn_cw)] >= score + 1001)
         ) or is_end(new_pos):
        new_path = path.copy()
        new_path.append(new_pos)
        ret.append((new_pos, turn_cw, new_path, score + 1001))
        
    turn_ccw = (-dir[1], dir[0])
    new_pos = (pos[0] + turn_ccw[0], pos[1] + turn_ccw[1])
    if (is_empty(new_pos, grid) and
        ((new_pos, turn_ccw) not in visited or visited[(new_pos, turn_ccw)] >= score + 1001)
         ) or is_end(new_pos):
        new_path = path.copy()
        new_path.append(new_pos)
        ret.append((new_pos, turn_ccw, new_path, score + 1001))
    
    return ret

dirs = [(0, 1), (-1, 0), (0, -1), (1, 0)]

queue = deque([(start, dirs[0], [start], 0)])
visited = {}
best = 1e9
paths = []

while len(queue) > 0:
    pos, dir, path, score = queue.pop()
    visited[(pos, dir)] = score
    
    if is_end(pos):
        print(f"End found: {score}")
        if score == best:
            print("Same score")
            paths.append(path)
        if score < best:
            best = score
            paths = [path]
    else:
        if score < best:
            queue.extend(walk(pos, dir, path, score, grid, visited))
    
unique_pos = set()
for path in paths:
    for pos in path:
        unique_pos.add(pos)
    
best, len(unique_pos)
```
